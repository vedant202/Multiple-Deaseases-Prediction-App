```python
# filling age null values with median
def filling_age(kd):
    rows = kd.shape[0]
    for row in range(0,rows):
        age = kd['age'][row]
        cate = kd['classification'][row]
#         'ckd':1,'notckd':0

        if(pd.isnull(age) and cate==1.0):
            kd['age'][row] = 58
        elif(pd.isnull(age) and cate==0.0):
            kd['age'][row] = 47
    return kd
```

```python
import matplotlib.pyplot as plt

def filling_column(kd2,column):
    col_type = kd2[column].dtypes
    print("Type of column is ",col_type)
    count_null_values = kd2[column].isnull().sum()
    print(f"checking null values in {column} column ",kd2[column].isnull().sum())
    sns.scatterplot(data=kd2,x='classification',y=column)
    plt.show()
    sns.boxplot(data=kd2,x='classification',y=column)
    plt.show()
    print(kd2[column].describe())
    
    ls_med_cla_0 = np.empty((0))
    ls_med_cla_1 = np.empty((0))


    for i in range(0,kd2[column].shape[0]):
        column_name = kd2[column][i]
        if(not pd.isnull(column_name)):
            if(kd2['classification'][i] == 0.0):
                ls_med_cla_0 = np.append(ls_med_cla_0,column_name)
            else:
                ls_med_cla_1 = np.append(ls_med_cla_1,column_name)
    # median value for column_name for 0 label
    print(f"checking median value for {column} for 0 label",np.median(ls_med_cla_0))
    print(f"checking median value for {column} for 1 label",np.median(ls_med_cla_1))
    
    median_value_0 = np.median(ls_med_cla_0)
    median_value_1 = np.median(ls_med_cla_1)
    
    for i in range(0,kd2[column].shape[0]):
        column_value = kd2[column][i]
        cla = kd2['classification'][i]
        
        if(pd.isnull(column_value) and cla ==0.0):
            kd2[column][i] = median_value_0
        elif(pd.isnull(column_value) and cla ==1.0):
            kd2[column][i] = median_value_1
            
    return kd2[column]

    
```

```python
kd2['sc'] = filling_column(kd2,"sc")
```
