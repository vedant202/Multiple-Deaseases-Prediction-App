```python
# filling age null values with median
def filling_age(kd):
    rows = kd.shape[0]
    for row in range(0,rows):
        age = kd['age'][row]
        cate = kd['classification'][row]
#         'ckd':1,'notckd':0

        if(pd.isnull(age) and cate==1.0):
            kd['age'][row] = 58
        elif(pd.isnull(age) and cate==0.0):
            kd['age'][row] = 47
    return kd
```

```python
kd2 = filling_age(kd1) 
kd2['age'].isnull().sum()
```

```python
sns.boxplot(data=kd2,x='classification',y='bp')
```

```python
def filling_bp(kd):
    rows = kd.shape[0]
    for row in range(0,rows):
        bp = kd['bp'][row]
        cla = kd['classification'][row]
        
        if(pd.isnull(bp) and cla==1.0):
            kd['bp'][row] = 80
        elif(pd.isnull(bp) and cla==0.0):
            kd['bp'][row] = 70
    return kd
```

```python
kd2 = filling_bp(kd1) 
print(kd2['age'].isnull().sum(),kd2['bp'].isnull().sum())
```

```python
def filling_pcv(kd2):
    rows = kd2.shape[0]
    for row in range(0,rows):
        pcv = kd2['pcv'][row]
        cl = kd2['classification'][row]
        
        if(pd.isnull(pcv) and cl == 1.0):
            kd2['pcv'][row] = 33
        elif(pd.isnull(pcv) and cl == 0.0):
            kd2['pcv'][row] = 46
    return kd2['pcv']
            
```
